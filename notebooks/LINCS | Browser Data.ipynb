```python
# Read data
lincs_dataset_dataframe = pd.DataFrame(json.loads(unicode(urllib2.urlopen('http://lincsportal.ccs.miami.edu/dcic/api/fetchdata?searchTerm=*&limit=2000').read(), errors='ignore'))['results']['documents']).set_index('datasetid')
lincs_dataset_dataframe.head()
```

```python
lincs_dataset_dataframe.columns
```

```python
# Filter
dataset_dataframe = lincs_dataset_dataframe[['datasetgroup', 'ldplink', 'technologies', 'cellline', 'smallmolecule']].reset_index().dropna()
dataset_dataframe.head()
```

```python
# Save
dataset_dataframe.to_csv('data/lincs_datasets.txt', index=False)
```
